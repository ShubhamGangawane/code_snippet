# 3.Reading Tabular Datafile in Pandas

```python
import pandas as pd
```

### Pandas has built-in function for reading different tabular datafile.

### pandas.read_table(file,sep=)

pd.read_table() is very fexible function which is used to read also every filetype given a seperater.
It takes filename with extension or path of the file and the sep parameter is use to define the type of file
eg. 

- sep=',' for csv (comma seperated values)
- sep='|' for psv (pipe seperated values)
- sep='\t' for tsv (tab seperated values)

Note: pd.read_table() has sep='\t' (tsv) as default seperater.


## a.TSV file

```python
df = pd.read_table('Datasets/chiporders.tsv')     #Since tsv is a default seperator.
df.head()                                         #head() is used to display 1st 5 row in table.
```

## b.CSV file

```python
#Method 1
df = pd.read_table('Datasets/sales.csv',sep=',')
df.head()
```

```python
#Method 2
df = pd.read_csv('Datasets/sales.csv')          #pd.read_csv() is ment to read csv file
df.head()
```

## d.PSV file

### As you can see this dataframe has no header or no column name, so let's add a suitable header to this.

## Adding Header

```python
colname = ['user srno','user age','user gender','user occupation','user zipcode']
df = pd.read_table('Datasets/psd.user',sep='|',header=None,names=colname)
df.head()
```
